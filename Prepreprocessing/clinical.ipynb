```python
import pandas as pd
import re

# label에서 d~ 를 추출한 코드
def extract_days_from_entry(label):
    match = re.search(r'd(\d+)$', label)
    return int(match.group(1)) if match else None

#'T1w'와 'bold-rest' 스캔을 모두 가진 환자 식별
T1w_subjects = set(OASIS3T[OASIS3T['scan category'] == 'T1w']['subject_id'])
bold_subjects = set(OASIS3T[OASIS3T['scan category'] == 'bold-rest']['subject_id'])
valid_subjects = T1w_subjects.intersection(bold_subjects)

OASIS3T_filtered = OASIS3T[
    (OASIS3T['subject_id'].isin(valid_subjects)) & 
    (OASIS3T['scan category'].isin(['T1w', 'bold-rest']))
]

# 3. 'days from entry' 추출
OASIS3T_filtered['days_from_entry'] = OASIS3T_filtered['label'].apply(extract_days_from_entry)

OASIS3T_latest = OASIS3T_filtered.sort_values('days_from_entry').groupby(['subject_id', 'scan category']).last().reset_index()

OASIS3T_wide = OASIS3T_latest.pivot(index='subject_id', columns='scan category', values=OASIS3T_latest.columns.drop(['subject_id', 'scan category']))
OASIS3T_wide.columns = [f'{col[0]}_{col[1]}' for col in OASIS3T_wide.columns]
OASIS3T_wide = OASIS3T_wide.reset_index()

clinical_data = [ADRC, udsb4, udsa1, udsa2, udsc1, udsb8, udsb1]
merged_data = OASIS3T_wide.copy()
```

```python
import pandas as pd
import numpy as np
import re

for data in clinical_data:
    if 'OASISID' in data.columns:
        merge_key = 'OASISID'
    elif 'Subject' in data.columns:
        merge_key = 'Subject'
    else:
        print(f"Skipping dataset: {data}, no matching ID column found")
        continue
    
    if 'days from entry' in data.columns:
        days_column = 'days from entry'
    elif 'days_to_visit' in data.columns:
        days_column = 'days_to_visit'
    else:
        print(f"Skipping dataset: {data}, no 'days from entry' or 'days_to_visit' column found")
        continue

    for scan_type in ['T1w', 'bold-rest']:
        temp_merge = pd.merge(merged_data, data, left_on='subject_id', right_on=merge_key, how='left')
        
        temp_merge[f'days_from_entry_{scan_type}'] = pd.to_numeric(temp_merge[f'days_from_entry_{scan_type}'], errors='coerce')
        temp_merge[days_column] = pd.to_numeric(temp_merge[days_column], errors='coerce')
        
        temp_merge['date_diff'] = np.abs(temp_merge[f'days_from_entry_{scan_type}'] - temp_merge[days_column])
        temp_merge = temp_merge[temp_merge['date_diff'] <= 180]  # 6개월(약 180일) 이내 데이터만 유지
        
        temp_merge = temp_merge.dropna(subset=['date_diff'])
        
        if not temp_merge.empty:
            temp_merge = temp_merge.loc[temp_merge.groupby('subject_id')['date_diff'].idxmin()]
        
        columns_to_merge = [col for col in data.columns if col not in [merge_key, days_column]]
        columns_to_merge = [col for col in columns_to_merge if col in temp_merge.columns] 
        
        if columns_to_merge and not temp_merge.empty:  
            merged_data = pd.merge(merged_data, temp_merge[['subject_id'] + columns_to_merge], 
                                   on='subject_id', how='left', suffixes=('', f'_{scan_type}'))
        else:
            print(f"No columns to merge or no matching data for {scan_type} in dataset {data}")

# 중복 컬럼 처리 및 필요없는 컬럼 제거

print(merged_data.head())
print(merged_data.columns)
```

```python

if 'CDRTOT' in merged_data.columns:
    cdr_distribution = merged_data['CDRTOT'].value_counts().sort_index()
    print("\nCDR Distribution:")
    print(cdr_distribution)
    
    cdr_percentage = cdr_distribution / len(merged_data) * 100
    print("\nCDR Distribution (%):")
    print(cdr_percentage)
    
    import matplotlib.pyplot as plt
    plt.figure(figsize=(10, 6))
    cdr_distribution.plot(kind='bar')
    plt.title('CDR Distribution')
    plt.xlabel('CDR Score')
    plt.ylabel('Count')
    plt.xticks(rotation=0)
    plt.show()
else:
    print("CDRTOT column not found in the merged data.")
```
