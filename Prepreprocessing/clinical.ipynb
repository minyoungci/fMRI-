```python
import pandas as pd
import numpy as np
import re

# GitHub 방식의 날짜 추출 함수 수정
def extract_days_from_entry(label):
    match = re.search(r'd(\d{4})', label)
    if match:
        return int(match.group(1))
    return np.nan

OASIS3T = OASIS3T.copy()
OASIS3T['days_from_entry'] = OASIS3T['label'].apply(extract_days_from_entry)

T1w_subjects = set(OASIS3T[OASIS3T['scan category'] == 'T1w']['subject_id'])
bold_subjects = set(OASIS3T[OASIS3T['scan category'] == 'bold-rest']['subject_id'])
valid_subjects = T1w_subjects.intersection(bold_subjects)

# 'label' 컬럼을 포함하여 최신 MRI 데이터 선택
latest_mri_dates = OASIS3T[OASIS3T['subject_id'].isin(valid_subjects)] \
    .sort_values('days_from_entry') \
    .groupby('subject_id').last().reset_index()[['subject_id', 'days_from_entry', 'label']]

# 임상 데이터 처리 함수
def process_clinical_data(data, id_column, date_column):
    data = data.copy()
    if date_column not in data.columns:
        data[date_column] = 0  # 날짜 정보가 없는 경우 0으로 설정
    data['days_from_entry'] = data[date_column].apply(extract_days_from_entry)
    return data.sort_values('days_from_entry').groupby(id_column).last().reset_index()

ADRC_processed = process_clinical_data(ADRC, 'Subject', 'ADRC_ADRCCLINICALDATA ID')
udsb4_processed = process_clinical_data(udsb4, 'OASISID', 'OASIS_session_label')
udsa1_processed = process_clinical_data(udsa1, 'OASISID', 'OASIS_session_label')
udsa2_processed = process_clinical_data(udsa2, 'OASISID', 'OASIS_session_label')
udsc1_processed = process_clinical_data(udsc1, 'OASISID', 'OASIS_session_label')
udsb8_processed = process_clinical_data(udsb8, 'OASISID', 'OASIS_session_label')
udsb1_processed = process_clinical_data(udsb1, 'OASISID', 'OASIS_session_label')

final_data = latest_mri_dates.rename(columns={'subject_id': 'OASISID'})

for name, data in [('ADRC', ADRC_processed), ('udsb4', udsb4_processed), ('udsa1', udsa1_processed),
                   ('udsa2', udsa2_processed), ('udsc1', udsc1_processed), ('udsb8', udsb8_processed),
                   ('udsb1', udsb1_processed)]:
    id_column = 'Subject' if name == 'ADRC' else 'OASISID'
    columns = [col for col in columns_to_merge[name] if col in data.columns]
    final_data = pd.merge(final_data, data[columns], left_on='OASISID', right_on=id_column, how='left')
    if id_column in final_data.columns and id_column != 'OASISID':
        final_data = final_data.drop(columns=[id_column])

final_data = final_data.rename(columns={'OASISID': 'subject_id'})

if 'cdr' in final_data.columns:
    original_count = len(final_data)
    final_data = final_data.dropna(subset=['cdr'])
    removed_count = original_count - len(final_data)
    print(f"\nRemoved {removed_count} rows with NaN values in 'cdr' column.")
    print(f"Remaining rows: {len(final_data)}")
else:
    print("\n'cdr' column not found in the dataset.")

print("\nFinal data shape:", final_data.shape)
print("Number of unique subjects:", final_data['subject_id'].nunique())
print("Columns in final data:", final_data.columns)

for col in ['cdr', 'mmse']:
    if col in final_data.columns:
        print(f"\n{col} column exists in final_data")
        print(f"{col} column data types:", final_data[col].dtype)
        print(f"{col} column unique values:", final_data[col].unique())
    else:
        print(f"\n{col} column does not exist in final_data")
        print(f"Columns that might be related to {col}:", [c for c in final_data.columns if col in c.lower()])

print("\nFinal data sample:")
print(final_data.head())
```
