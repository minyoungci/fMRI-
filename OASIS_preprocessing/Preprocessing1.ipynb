```python
import os
import nibabel as nib
import matplotlib.pyplot as plt

def visualize_nifti_files(directory):
    # 디렉토리 내의 모든 .nii.gz 파일 찾기
    nifti_files = [file for file in os.listdir(directory) if file.endswith('.nii.gz')]

    # 각 파일 시각화
    for file in nifti_files:
        # NIfTI 파일 로드
        img = nib.load(os.path.join(directory, file))
        data = img.get_fdata()

        print(f"File: {file}, Shape: {data.shape}")

        # 데이터 형상에 따라 중간 슬라이스 선택
        if len(data.shape) == 4:
            mid_slice = data.shape[2] // 2
            data_slice = data[:, :, mid_slice, 0]
        elif len(data.shape) == 3:
            mid_slice = data.shape[2] // 2
            data_slice = data[:, :, mid_slice]
        else:
            print(f"Unexpected data shape for file: {file}")
            continue

        # 선택한 슬라이스 시각화
        plt.figure()
        plt.imshow(data_slice, cmap='gray')
        plt.title(file)
        plt.axis('off')

    plt.tight_layout()
    plt.show()

directory='/home/minyoungxi/MINYOUNGXI/XFL/OASIS_615_fmriprep_output/sub-OAS30078/ses-d0210/func' # 이 부분에 실제 디렉토리 경로를 입력하세요.
visualize_nifti_files(directory)
```

```python
import os
import numpy as np
import pandas as pd
import nibabel as nib
from nilearn.image import resample_to_img
from nilearn.input_data import NiftiMasker
from nilearn.plotting import plot_roi
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

def _resample_and_mask(func_file, mask_file, **kwargs):
   resampled_func_file = resample_to_img(source_img=func_file, target_img=mask_file, interpolation='continuous')

   masker = NiftiMasker(mask_img=mask_file, **kwargs)
   masked_func_file = masker.fit_transform(resampled_func_file)
   return masked_func_file

def _gen_run_list(base_dir):
   run_dict = {}

   for dir_path, dir_names, file_names in os.walk(base_dir):
       for file_name in [f for f in file_names if f.endswith('nii.gz')]:
           session = dir_path.split('/')[9]

           if session not in run_dict:
               run_dict[session] = []

           run_dict[session].append(os.path.join(dir_path, file_name))

   run_df = pd.concat({k: pd.Series(v) for k, v in run_dict.items()}, axis=1)
   run_df = run_df.melt(var_name='session', value_name='func_file').dropna()

   return run_df

def _get_runs_bids_file(func_paths, mask_paths=None):
   func_df = pd.DataFrame(func_paths, columns=['func_file'])
   func_df['session'] = func_df['func_file'].str.split('/').str[8]
   func_df['run'] = func_df['func_file'].str.split('_').str[3]
   func_df['run_id'] = func_df['run'].str.extract('(\d+)', expand=True).astype(int)

   if mask_paths:
       mask_df = pd.DataFrame(mask_paths, columns=['mask_file'])
       func_df = pd.concat([func_df, mask_df], axis=1)
   return func_df

def preprocess_oasis(data_dir):
   out_dir = '/'.join(data_dir.split('/')[:-1])
   out_dir_npy = f"{out_dir}/fmriprep_masked/sub-{data_dir.split('/')[-1]}"
   os.makedirs(out_dir_npy, exist_ok=True)

   run_df = _gen_run_list(data_dir)
   func_file_list = run_df['func_file'].tolist()
   mask_file = f"{data_dir}/anat/sub-{data_dir.split('/')[-1]}_space-MNI152NLin2009cAsym_desc-brain_mask.nii.gz"
   kwargs = {'t_r': 2.2, 'smoothing_fwhm': None, 'standardize': True, 'detrend': True, 'low_pass': 0.08, 'high_pass': 0.01} 

   for i, func_file in enumerate(func_file_list):
       print(f"Preprocessing {i+1}/{len(func_file_list)}")
       masked_func_file = _resample_and_mask(func_file, mask_file, **kwargs)
       session = func_file.split('/')[8]
       file_name = func_file.split('/')[10].split('.')[0]
       file_name_npy = f"{out_dir_npy}/{session}/{file_name}.npy"
       os.makedirs(f"{out_dir_npy}/{session}", exist_ok=True)
       np.save(file_name_npy, masked_func_file)

   print("Getting run lists")
   npy_files = [f for f in os.listdir(out_dir_npy) if f.endswith('.npy')] 
   run_files_npy = [f"{out_dir_npy}/{f}" for f in npy_files]
   mask_files_npy = [f"{out_dir_npy}/anat/{f.split('_')[0]}_space-MNI152NLin2009cAsym_desc-brain_mask.npy" for f in npy_files]
   run_df_npy = _get_runs_bids_file(run_files_npy, mask_files_npy)
   run_df_npy = run_df_npy.sort_values(by=['session', 'run_id'])

   out_file_npy = f"{out_dir}/bids_run_list_npy.pkl"
   run_df_npy.to_pickle(out_file_npy)
   print(f"Run list saved to {out_file_npy}")
```
