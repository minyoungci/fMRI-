```python
import os
import nibabel as nib
import matplotlib.pyplot as plt

def visualize_nifti_files(directory):
    #    .nii.gz  
    nifti_files = [file for file in os.listdir(directory) if file.endswith('.nii.gz')]

    #   
    for file in nifti_files:
        # NIfTI  
        img = nib.load(os.path.join(directory, file))
        data = img.get_fdata()

        #   
        mid_slice = data.shape[2] // 2

        #   
        plt.figure()
        plt.imshow(data[:, :, mid_slice], cmap='gray')
        plt.title(file)
        plt.axis('off')

    plt.tight_layout()
    plt.show()

directory='/home/minyoungxi/MINYOUNGXI/XFL/OASIS_615_fmriprep_output/sub-OAS30078/ses-d0210/anat' #      .
visualize_nifti_files(directory)
```
