```python
import os
import nibabel as nib
import matplotlib.pyplot as plt

def visualize_nifti_files(directory):
    # 디렉토리 내의 모든 .nii.gz 파일 찾기
    nifti_files = [file for file in os.listdir(directory) if file.endswith('.nii.gz')]

    # 각 파일 시각화
    for file in nifti_files:
        # NIfTI 파일 로드
        img = nib.load(os.path.join(directory, file))
        data = img.get_fdata()

        # 중간 슬라이스 선택
        mid_slice = data.shape[2] // 2

        # 선택한 슬라이스 시각화
        plt.figure()
        plt.imshow(data[:, :, mid_slice], cmap='gray')
        plt.title(file)
        plt.axis('off')

    plt.tight_layout()
    plt.show()

directory='/home/minyoungxi/MINYOUNGXI/XFL/OASIS_615_fmriprep_output/sub-OAS30078/ses-d0210/anat' # 이 부분에 실제 디렉토리 경로를 입력하세요.
visualize_nifti_files(directory)
```
