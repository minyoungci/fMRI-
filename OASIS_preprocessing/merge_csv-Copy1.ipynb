### 알츠하이머 관련 컬럼

- NORMCOG: 정상 인지 상태 여부. 0이 아닌 경우 인지 장애를 가지고 있을 가능성이 있음. 

- DEMENTED: 치매 여부를 나타냅니다.

- PROBAD: 알츠하이머병으로 진단된 경우.

- PROBADIF: 알츠하이머병 진단의 확실성.

- POSSAD: 가능성 있는 알츠하이머병.

- POSSADIF: 가능성 있는 알츠하이머병 진단의 확실성.pd.set_option('display.max_columns', None)
